```python

import basedosdados as bd
import pandas as pd

# Para carregar o dado direto no pandas
import basedosdados as bd

# Para carregar o dado direto no pandas
df = bd.read_table(dataset_id='br_poder360_pesquisas',
table_id='microdados',
billing_project_id="databaelection")
df.head()                                #para visualizar os dados
```

Atribuição da linha de código de seleção a uma nova variável  presidente2022: facilitar manipulação

selecão de várias colunas usando o boolean & e selecionando de uma das colunas o uum nome ou o outro e ordenado por datas)

```python
eleicao_2022 = df[(df.instituto=='Datafolha')
 & (df.ano==2022) & (df.cargo =='presidente') 
 & (df.sigla_uf.isnull()) & (df.tipo=='estimulada') 
 & (df.turno) & ((df.nome_candidato == 'Bolsonaro') | (df.nome_candidato =='Lula')) 
 &  (df.descricao_cenario.str.contains('cenário 1') == True)].sort_values(by="data")
 
eleicao_2022.head()

```
